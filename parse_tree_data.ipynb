```python
import pandas as pd
import csv
import numpy as np
```

```python
first_data = pd.read_csv('2005_Street_Tree_Census.csv')
second_data = pd.read_csv('2015_Street_Tree_Census.csv')
```

```python
first_data
```

```python
# Group trees into geospatial segments by lat and long
step = 0.00000001 # This should be about 1/4-1/2 a city block 
to_bin = lambda x: np.floor(x / step) * step
first_data["latBin"] = to_bin(first_data.latitude)
first_data["lonBin"] = to_bin(first_data.longitude)
groups = first_data.groupby(["latBin", "lonBin"])
```
